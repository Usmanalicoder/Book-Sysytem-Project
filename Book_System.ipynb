**CSV Files change into Json Files**

```python
import csv
import json

# Function to load data from a CSV file and convert it to a JSON file
def csv_to_json(input_csv, output_json):
    data = []
    with open(input_csv, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            data.append(row)

    with open(output_json, 'w') as json_file:
        json.dump(data, json_file, indent=4)

# Load data from CSV files and convert to JSON
csv_to_json("books.csv", "books.json")
csv_to_json("bookloans.csv", "book_loans.json")
```

```python
books_data = [
    {
        'book_number': 1,
        'book_title': 'Fundamentals Of Wavelets',
        'author': 'Jaideva Goswami',
        'borrowed_date': '2022-12-31',
        'return_date': '2023-01-10',
    },
    {
        'book_number': 2,
        'book_title': 'Data Smart',
        'author': 'John Foreman',
        'borrowed_date': '2023-01-01',
        'return_date': '2023-01-15',
    },
    {
        'book_number': 3,
        'book_title': 'God Created The Integers',
        'author': 'Stephen Hawking',
        'borrowed_date': '2023-02-01',
        'return_date': '2023-02-15',
    },
    {
        'book_number': 4,
        'book_title': 'Superfreakonomics',
        'author': 'Stephen Dubner',
        'borrowed_date': '2023-03-01',
        'return_date': '2023-03-10',
    },
    # Add more books as needed
]

# Function to calculate the number of days between two dates
def calculate_days_between_dates(start_date, end_date):
    start_date = datetime.strptime(start_date, '%Y-%m-%d')

    if end_date:
        end_date = datetime.strptime(end_date, '%Y-%m-%d')
        delta = end_date - start_date
        return delta.days
    else:
        return None

# Function to calculate borrowing statistics
def calculate_borrowing_statistics(books):
    book_statistics = {}

    for book in books:
        days_borrowed = calculate_days_between_dates(book['borrowed_date'], book['return_date'])
        year_borrowed = datetime.strptime(book['borrowed_date'], '%Y-%m-%d').year

        # Update overall statistics
        if book['book_number'] not in book_statistics:
            book_statistics[book['book_number']] = {
                'total_borrowings': 0,
                'total_days_borrowed': 0,
                'days_borrowed_2023': 0
            }

        book_statistics[book['book_number']]['total_borrowings'] += 1
        book_statistics[book['book_number']]['total_days_borrowed'] += days_borrowed

        # Update statistics for the year 2023
        if year_borrowed == 2023:
            book_statistics[book['book_number']]['days_borrowed_2023'] += days_borrowed

    return book_statistics

# Display borrowing statistics
book_statistics = calculate_borrowing_statistics(books_data)

# Print statistics for each book
for book_number, statistics in book_statistics.items():
    print(f"Book Number: {book_number}")
    print(f"Total Borrowings: {statistics['total_borrowings']}")
    print(f"Total Days Borrowed: {statistics['total_days_borrowed']}")
    print(f"Days Borrowed in 2023: {statistics['days_borrowed_2023']}")
    print("-" * 30)
```

```python
#structure with genre information
books_data = [
    {
        'book_number': 1,
        'book_title': 'Fundamentals Of Wavelets',
        'author': 'Jaideva Goswami',
        'genre': 'Tech',
        'borrowed_date': '2022-12-31',
        'return_date': '2023-01-10',
    },
    {
        'book_number': 2,
        'book_title': 'Data Smart',
        'author': 'John Foreman',
        'genre': 'Tech',
        'borrowed_date': '2023-01-01',
        'return_date': '2023-01-15',
    },
    {
        'book_number': 3,
        'book_title': 'God Created The Integers',
        'author': 'Stephen Hawking',
        'genre': 'Tech',
        'borrowed_date': '2023-02-01',
        'return_date': '2023-02-15',
    },
    {
        'book_number': 4,
        'book_title': 'Superfreakonomics',
        'author': 'Stephen Dubner',
        'genre': 'Science',
        'borrowed_date': '2023-03-01',
        'return_date': '2023-03-10',
    },
    # Add more books as needed
]

# Function to calculate the number of days between two dates
def calculate_days_between_dates(start_date, end_date):
    start_date = datetime.strptime(start_date, '%Y-%m-%d')

    if end_date:
        end_date = datetime.strptime(end_date, '%Y-%m-%d')
        delta = end_date - start_date
        return delta.days
    else:
        return None

# Function to calculate popularity statistics for genres
def calculate_genre_statistics(books):
    genre_statistics = {}

    for book in books:
        days_borrowed = calculate_days_between_dates(book['borrowed_date'], book['return_date'])
        year_borrowed = datetime.strptime(book['borrowed_date'], '%Y-%m-%d').year

        # Update overall genre statistics
        if book['genre'] not in genre_statistics:
            genre_statistics[book['genre']] = {
                'total_borrowings': 0,
                'total_days_borrowed': 0,
                'books_count': 0
            }

        genre_statistics[book['genre']]['total_borrowings'] += 1
        genre_statistics[book['genre']]['total_days_borrowed'] += days_borrowed
        genre_statistics[book['genre']]['books_count'] += 1

        # Update genre statistics for the year 2023
        if year_borrowed == 2023:
            if 'days_borrowed_2023' not in genre_statistics[book['genre']]:
                genre_statistics[book['genre']]['days_borrowed_2023'] = 0

            genre_statistics[book['genre']]['days_borrowed_2023'] += days_borrowed

    return genre_statistics

# Display genre statistics
genre_statistics = calculate_genre_statistics(books_data)

# Print statistics for each genre
print("General Popularity of Genres over the Years:")
for genre, statistics in genre_statistics.items():
    print(f"Genre: {genre}")
    print(f"Total Borrowings: {statistics['total_borrowings']}")
    print(f"Total Days Borrowed: {statistics['total_days_borrowed']}")
    print(f"Number of Books: {statistics['books_count']}")
    print("-" * 30)

# Print statistics for each genre in the year 2023
print("\nPopularity of Genres in 2023:")
for genre, statistics in genre_statistics.items():
    if 'days_borrowed_2023' in statistics:
        print(f"Genre: {genre}")
        print(f"Days Borrowed in 2023: {statistics['days_borrowed_2023']}")
        print("-" * 30)

# Find the most popular genre(s)
most_popular_genres = max(genre_statistics, key=lambda k: genre_statistics[k]['total_borrowings'])
most_popular_genre_count = genre_statistics[most_popular_genres]['books_count']

print(f"\nMost Popular Genre(s): {most_popular_genres} ({most_popular_genre_count} books)")
```

```python
def calculate_overall_statistics(books):
    total_loans = len(books)
    total_days_borrowed = sum(calculate_days_between_dates(book['borrowed_date'], book['return_date']) for book in books if book['return_date'])

    late_books = [book for book in books if book['return_date'] and calculate_days_between_dates(book['borrowed_date'], book['return_date']) > 0]
    loans_returned_late = len(late_books)
    total_days_late = sum(calculate_days_between_dates(book['borrowed_date'], book['return_date']) for book in late_books)

    average_days_borrowed = total_days_borrowed / total_loans if total_loans > 0 else 0
    average_days_late = total_days_late / loans_returned_late if loans_returned_late > 0 else 0

    return {
        'total_loans': total_loans,
        'total_days_borrowed': total_days_borrowed,
        'loans_returned_late': loans_returned_late,
        'total_days_late': total_days_late,
        'average_days_borrowed': average_days_borrowed,
        'average_days_late': average_days_late,
    }
```

```python
# Display overall borrowing statistics
overall_statistics = calculate_overall_statistics(books_data)
```
